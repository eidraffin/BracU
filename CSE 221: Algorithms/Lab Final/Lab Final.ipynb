```python
import heapq

f1 = open('input.txt', 'r')

lines = f1.readlines()
data = [tuple(map(str, i.split())) for i in lines]


graph = {}
for edge in data:
    if edge[0] not in graph:
        graph[edge[0]] = []
    if edge[1] not in graph:
        graph[edge[1]] = []

    graph[edge[0]].append((int(edge[2]), edge[1]))
    graph[edge[1]].append((int(edge[2]), edge[0]))

#print(graph)
print("Adjacency List:\n")
for k, v in graph.items():
    print(f"{k}: {v}")
print("\n------------------------------------------------------------------------------------------------------\n")


def dijkstra(graph, start, end):
    priority_queue = [(0, start, [])]
    visited = set()

    while priority_queue:
        cost, node, path = heapq.heappop(priority_queue)
        if node not in visited:
            visited.add(node)
            path = path + [node]
            if node == end:
                return cost, path

            for c, neighbour in graph[node]:
                if neighbour not in visited:
                    heapq.heappush(priority_queue, (cost + int(c), neighbour, path))

    return float("inf"), []


# Task A (showing Washington to Kentucky)
start = "Washington"
end = "Kentucky"

cost, path = dijkstra(graph, start, end)

print(f"Shortest distance: {str(cost)} km")
print(f"Route: {' '.join(path)}")
print()


# Task B (starting from Kentucky and Florida)
start_a = "Kentucky"
start_b = "Florida"
end_a = "Florida"
end_b = "Kentucky"

cost_a, path_a = dijkstra(graph, start_a, end_a)
cost_b, path_b = dijkstra(graph, start_b, end_b)

common_node = None
common_cost = float("inf")

for node in path_a:
    if node in path_b:
        common_node = node
        common_cost = min(common_cost, cost_a + cost_b)
        break

meeting = ''
for i in path_a[1:-1]:
    if i in path_b:
        meeting = i

print("The meeting district:", meeting)
print("Route for Adam:", " ".join(path_a[0:-1]))
print("Route for Lisa:", " ".join(path_b[0:-1]))


f2 = open('output.txt', 'w')

f2.write(f"Shortest distance: {str(cost)} km\n")
f2.write(f"Route: {' '.join(path)}")
f2.write("\n\n")
f2.write(f"The meeting district: {meeting}\n")
f2.write(f"Route for Adam: {' '.join(path_a[0:-1])}\n")
f2.write(f"Route for Lisa: {' '.join(path_b[0:-1])}")

f1.close()
f2.close()
```
