```python
#Task5

def DomainReplace(email, new, old='kaaj.com'):
    find = email.index("@")

    if email[find+1:] == new:
        return f"Unchanged: {email}"
    elif email[find+1:] == old:
        email = email[:find+1] + new
        return f"Changed: {email}"

print(DomainReplace('alice@kaaj.com', 'sheba.xyz'))
print(DomainReplace('bob@sheba.xyz', 'sheba.xyz'))
```

```python
#Task9

def capitalization_fix(string):
	fixed_string = str()
	if 90 <= ord(string[0]) <= 122:
		fixed_string += chr(ord(string[0]) - 32)
	else:
		fixed_string += string[0]
	i = 1
	while i < len(string):
		if string[i] in '.?!' and i != len(string) - 1:
			fixed_string += string[i] + ' '
			i += 2
			if 90 <= ord(string[i]) <= 122:
				fixed_string += chr(ord(string[i]) - 32)
				i += 1
			else:
				fixed_string += string[i]
				i += 1
		elif string[i-1] == ' ' and string[i] == 'i' and string[i+1] == ' ':
			fixed_string += 'I'
			i += 1
		else:
			fixed_string += string[i]
			i += 1
	return fixed_string


print(capitalization_fix("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog's name? i love my pet very much."))
```

```python
#Task2

class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)


class ComplexNumber(RealNumber):
    def __init__(self, r=0, i=0):
        if type(r) == int:
            super().__init__(r)
        else:
            super().__init__(r.number)
        self.imaginaryNumber = i
    def __add__(self, anotherComplexNumber):
        r = super().__add__(anotherComplexNumber)
        i = self.imaginaryNumber + anotherComplexNumber.imaginaryNumber
        return ComplexNumber(r, i)
    def __sub__(self, anotherComplexNumber):
        r = super().__sub__(anotherComplexNumber)
        i = self.imaginaryNumber - anotherComplexNumber.imaginaryNumber
        return ComplexNumber(r, i)
    def __str__(self):
        if self.imaginaryNumber >= 0:
            return f"{super().__str__()} + {self.imaginaryNumber}i"
        else:
            return f"{super().__str__()} - {-self.imaginaryNumber}i"


r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#Task6

class Shape3D:
	
	pi = 3.14159
	def __init__(self, name = 'Default', radius = 0):
		self._area = 0
		self._name = name
		self._height = 'No need'
		self._radius = radius   

	def calc_surface_area(self):
		return 2 * Shape3D.pi * self._radius

	def __str__(self):
		return "Radius: "+str(self._radius)


class Sphere(Shape3D):
	def __init__(self, name='Sphere', radius=0):
		super().__init__(name, radius)
		print(f"Shape name: Sphere, Area Formula: 4 * pi * r * r")
	def calc_surface_area(self):
		self._area = super().calc_surface_area() * 2 * self._radius
	def __str__(self):
		return super().__str__()+", Height: "+str(self._height)+"\nArea: "+str(self._area)


class Cylinder(Shape3D):
	def __init__(self, name='Cylinder', radius=0, height=0):
		super().__init__(name, radius)
		self._height = height
		print(f"Shape name: Cylinder, Area Formula: 2 * pi * r * (r + h)")
	def calc_surface_area(self):
		self._area = super().calc_surface_area() * (self._radius + self._height)
	def __str__(self):
		return super().__str__()+", Height: "+str(self._height)+"\nArea: "+str(self._area)


sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```
