```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row, col= floor.shape

  for i in range(col):
    if i%2 != 0:
      for j in range(row - 1, -1, -1):
        if j%2 != 0:
          print(floor[j][i], end=' ')
      print()

    else:
      for j in range(0, row, 1):
        if j%2 == 0:
          print(floor[j][i], end=' ')
      print()


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
  district_row, district_col = district.shape
  fence = np.zeros((district_row + (depth*2), district_col + (depth*2)), dtype = int)

  for i in range(fence.shape[0]):
    for j in range(fence.shape[1]):
      fence[i][j] = 8

  for i in range(district_row):
    for j in range(district_col):
      fence[i + depth][j + depth] = district[i][j]

  return fence


depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  size = (len(clubroom) * (len(clubroom) - 1)) // 2
  # Calculating the total potential matchups by multiplying the number of players, excluding diagonal elements and self-matches, because we cannot count the strength of players against themselves.
  # Floor dividing by 2 to ensure each player's strength difference is recorded only once in the 'strength_diff' array. Also it  ensures that only unique player matchups.
  strength_diff = np.zeros(size, dtype=int)
  k = 0

  for i in range(len(clubroom)):
    for j in range(i+1, len(clubroom)):
      strength_diff[k] = clubroom[i][j] - clubroom[j][i]
      k += 1

  return strength_diff


clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  matrix = np.zeros((n, n), dtype=int)
  count = 1

  for i in range(n):
    matrix[i][0] = matrix[i][i] = count
    count += 1
    sum = i
    for j in range(1, i):
      sum += matrix[i-1][j]
      matrix[i][j] = sum
  return matrix

def print_without_none(matrix):
  for i in range(len(matrix)):
    for j in range(i+1):
      print(matrix[i][j], end="  ")
    print()


n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def moving_around(cmds):
  grid = np.full((7, 7), '.')
  row, col = 3,3
  grid[row][col] = "-"

  def position(i, row, col):  # movements of up, down, right, and left
    if i == 1 or i == 5 or i == 9:
      if col - 1 >= 0:
        col = col - 1

    elif i == 2 or i == 6 or i == 10:
      if col + 1 <= 6:
        col = col + 1
        grid[row][col] = "*"

    elif i == 3 or i == 7 or i == 11:
      if row - 1 >= 0:
        row = row - 1
        grid[row][col] = "*"

    elif i == 4 or i == 8 or i == 12:
      if row + 1 <= 6:
        row = row + 1
        grid[row][col] = "*"

    return np.array([row, col])

  for i in range(len(cmds)):

    if cmds[i] == 1 or cmds[i] == 2 or cmds[i] == 3 or cmds[i] == 4:  # upper left
      if row - 2 >= 0 and col - 2 >= 0:
        row, col = row - 2, col - 2
        matrix = position(cmds[i], row, col)
        row, col = matrix[0], matrix[1]
        grid[row][col] = "*"

    elif cmds[i] == 5 or cmds[i] == 6 or cmds[i] == 7 or cmds[i] == 8:  # upper right
      if row - 2 >= 0 and col + 2 <= 6:
        row, col = row - 2, col + 2
        matrix = position(cmds[i], row, col)
        row, col = matrix[0], matrix[1]
        grid[row][col] = "*"

    elif cmds[i] == 9 or cmds[i] == 10 or cmds[i] == 11 or cmds[i] == 12:  # lower left
      if row + 2 <= 6 and col - 2 >= 0:
        row, col = row + 2, col - 2
        matrix = position(cmds[i], row, col)
        row, col = matrix[0], matrix[1]
        grid[row][col] = "*"

  grid[row][col] = "/"
  return grid


cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
