```python
class DUMMY_HEAD:
    def __init__(self, data= None, next= None, prev= None):
        self.data, self.next, self.prev = data, next, prev

class PATIENT:
    def __init__(self, id, name, age, bloodgroup, next= None, prev= None):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = next
        self.prev = prev

class WAITING_ROOM_MANAGEMENT:
    flag = True
    def __init__(self):
        self.head = DUMMY_HEAD()
        self.tail = self.head

    def registerPatient(self, id, name, age, bloodgroup):
        patient = PATIENT(id, name, age, bloodgroup)
        self.tail.next = patient
        self.tail = patient
        patient.prev = self.tail
        self.tail.next = self.head
        self.head.prev = self.tail
        WAITING_ROOM_MANAGEMENT.flag = False
        print(f"Patient registered successfully...\nName: {name}")

    def servePatient(self):
        if self.head.next == None:
            print("No patients left to serve!")
            WAITING_ROOM_MANAGEMENT.flag = True
        else:
            print(f"Now serving: {self.head.next.name}")
            a = self.head.next.next
            rem = self.head.next
            if rem.next == self.head:
                self.head.next = None
                self.head.prev = None
                self.tail = self.head
                WAITING_ROOM_MANAGEMENT.flag = True
            else:
                self.head.next = a
                a.prev = self.head
            rem.next = None
            rem.prev = None
            rem = None

    def cancelAll(self):
        self.head.next = None
        self.head.prev = None
        self.tail = self.head
        print(f"All appointments cancelled successfully...")
        WAITING_ROOM_MANAGEMENT.flag = True

    def canDoctorGoHome(self):
        print("Can the doctor go home?")
        if WAITING_ROOM_MANAGEMENT.flag:
            print("YES")
        else:
            print("NO")

    def showAllPatient(self):
        if WAITING_ROOM_MANAGEMENT.flag:
            print("No patient IDs available for display!")
        else:
            ptr = self.head.next
            print('Patient List:')
            while ptr is not self.head:
                print(f"ID: {ptr.id}, Name: {ptr.name}")
                ptr = ptr.next

    def reverseTheLine(self):
        if WAITING_ROOM_MANAGEMENT.flag:
            print("No more queues left to reverse!")
        else:
            prev_node = None
            ptr = self.head
            first_node = self.head.next
            while ptr is not None:
                next_node = ptr.next
                ptr.next = prev_node
                ptr.prev = next_node
                prev_node = ptr
                ptr = next_node
            self.tail = first_node
            self.head.prev = self.tail
            print("Patient queue reversed successfully...")
```

```python
# TESTER CODE
def main():
    try:
        print("\n__WAITING ROOM MANAGEMENT SYSTEM__\n")
        wrm = WAITING_ROOM_MANAGEMENT()
        while True:
            print("Option[S]:")
            print("1. Add Patient?")
            print("2. Serve Patient?")
            print("3. Show All Patients?")
            print("4. Can Doctor Go Home?")
            print("5. Cancel All Appointment?")
            print("6. Reverse Patient Queue?")
            print("7. Exit")
            option = input("\nEnter a number [1, 2, 3, 4, 5, 6, 7]:\n>>> ")

            if option == "1":
                print("\nRegistering patient...")
                id = int(input("Enter Patient ID:\n>>> "))
                name = input("Enter Patient Name:\n>>> ")
                age = int(input("Enter Patient Age:\n>>> "))
                bloodgroup = input("Enter Blood Group:\n>>> ")
                wrm.registerPatient(id, name, age, bloodgroup)

            elif option == "2":
                wrm.servePatient()

            elif option == "3":
                wrm.showAllPatient()

            elif option == "4":
                wrm.canDoctorGoHome()

            elif option == "5":
                wrm.cancelAll()

            elif option == "6":
                wrm.reverseTheLine()

            elif option == "7":
                print("Adios!\n")
                break

            else:
                print("\nInvalid Option! Please select a valid option between (1 to 7).")
            print("\n=================================================\n")
    except:
        print(f"Something went wrong! Try again '_' \n")

if __name__ == "__main__":
    main()
```
