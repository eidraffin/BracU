```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def convert_mirror(root):
    if root == None:
        return

    n = BTNode(root.elem)
    n.left = convert_mirror(root.right)
    n.right = convert_mirror(root.left)

    return n


#DRIVER CODE
root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  40 20 60 10 30
print()

root2 = convert_mirror(root)
print('Mirrored Tree Inorder Traversal: ', end = ' ')
inorder(root2) #Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root, level={}, idx=0):
    if root == None:
        return

    if idx not in level.keys() or root.elem < level[idx]:
        level[idx] = root.elem

    smallest_level(root.left, level, idx + 1)
    smallest_level(root.right, level, idx + 1)

    return level


#DRIVER CODE
root = tree_construction([None, 4,9,2,3,-5,None,7])
print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print('Level Wise Smallest Value: ', end = ' ')
print(smallest_level(root)) #Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def inorder_predecessor(root, x):
    if root == None:
        return

    if x.elem > root.elem:
        right = inorder_predecessor(root.right, x)
        if right is not None:
            return right
        else:
            return root

    return inorder_predecessor(root.left, x)


#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
def LCA(root, x, y):
    if root == None:
        return

    if x < root.elem and y < root.elem:
        return LCA(root.left, x, y)

    if x > root.elem and y > root.elem:
        return LCA(root.right, x, y)

    return root


# DRIVER CODE
root = BTNode(15)
n2 = BTNode(10)
n3 = BTNode(25)
n4 = BTNode(8)
n5 = BTNode(12)
n6 = BTNode(20)
n7 = BTNode(30)
n8 = BTNode(6)
n9 = BTNode(9)
n10 = BTNode(18)
n11 = BTNode(22)

root.left = n2
root.right = n3

n2.left = n4
n2.right = n5

n3.left = n6
n3.right = n7

n4.left = n8
n4.right = n9

n6.left = n10
n6.right = n11

l1 = LCA(root, 6, 12)
print(l1.elem)

l2 = LCA(root, 20, 6)
print(l2.elem)

l3 = LCA(root, 18, 22)
print(l3.elem)

l4 = LCA(root, 20, 25)
print(l4.elem)

l5 = LCA(root, 10, 12)
print(l5.elem)
```

BONUS TASK 1

```python
def summ(root):
    if root == None:
        return 0
    return summ(root.left) + root.elem + summ(root.right)

def is_sum_tree(root):
    if (root == None) or (root.left == None and root.right == None):
        return True

    left_sum = summ(root.left)
    right_sum = summ(root.right)

    if (root.elem == left_sum + right_sum) and is_sum_tree(root.left) and is_sum_tree(root.right):
        return True

    return False


#DRIVER CODE
root = BTNode(26)
n2 = BTNode(10)
n3 = BTNode(3)
n4 = BTNode(4)
n5 = BTNode(6)
n6 = BTNode(3)

root.left = n2
root.right = n3

n2.left = n4
n2.right = n5

n3.right = n6

result = is_sum_tree(root)
print(result)
```

```python
def difference_between_sum(root, level=0):
    if root == None:
        return 0

    summ = root.elem if level % 2 != 0 else (-root.elem)

    left_sum = difference_between_sum(root.left, level + 1)
    right_sum = difference_between_sum(root.right, level + 1)

    return summ + left_sum + right_sum


#DRIVER CODE
root = BTNode(1)
n2 = BTNode(2)
n3 = BTNode(3)
n4 = BTNode(4)
n5 = BTNode(5)
n6 = BTNode(6)
n7 = BTNode(7)
n8 = BTNode(8)
root.left = n2
root.right = n3

n2.left = n4
n3.left = n5
n3.right = n6

n5.left = n7
n5.right = n8

result = difference_between_sum(root)
print(result)
```
